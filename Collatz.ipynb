## 3n+1問題 (コラッツ予想）
**3n+1問題** とは整数に関する数学の問題で、任意の正の整数`n`に対して、

1. `n`が偶数なら`2`で割る
2. `n`が奇数なら`3`をかけて`1`を足す

という操作を繰り返すと、どんな正の整数`n`でも最終的には`1`になる、という数学者ローター・コラッツさんが1937年に提唱した数学の予想です。これは現在でも数学的に証明されていません。最近ネットニュースでもちょっと話題になりました（https://prtimes.jp/main/html/rd/p/000000015.000037669.html）。

### Pythonで3n+1問題を実装するには
まず最初に任意の正の整数`n`を選びます。

次に、選んだ`n`に対して、上の操作1., 2.を繰り返していきます。
Pythonでは、操作を繰り返すための文法として`for`文と`while`文がありますが、ここでは`n`が`1`になるまで繰り返すという条件の下で繰り返しを続けますので、その場合は`while`文を使います。

```python
while 条件式:
(indent->) 繰り返し処理する内容
```

これは、条件式が満たさている間は処理を繰り返すという構文になっています。ここで、`indent`とは、空白のことです。Pythonでは4文字分空けるのが決まりです。条件式は、`n != 1`を使います。「`n`は`1`でない」という意味になります。

上の操作1., 2.を判定するには`if`文を使います。

```python
if 条件式:
(indent->) 条件式を満たした場合の処理
else:
(indent->) 条件式を満たさない場合の処理
```

のように書きます。

問題では`n`が偶数か奇数かを知りたいわけなので、条件式としては`n%2 == 0`を使います。これは、Pythonの文法で「`n`を`2`で割った余りは`0`である」という条件を表します。

これらをまとめると、

```python
while n != 1:
    if n%2 == 0:
        処理１
    else:
        処理２
    
    #途中結果を出力する
    print(n) 
```
のような感じになります。

### 解答例
どうしてもわからなければ以下を参考にしてみてください。正の整数`n`をいろいろ変えてみて、最終的に`n=1`となることを確認してみましょう。

````python
n=12
````

````python
while n != 1:
    if n%2 == 0:
        n = n//2
    else:
        n = 3*n+1
    
    print(n)
````

上で`//`は整数の除算を表します（実数の除算は`/`）。
`n`が変化する様子を可視化するとおもしろい（追加したプログラムの詳しい内容は１０月以降にやります）。

````python
n=22
result=[n]
while n != 1:
    if n%2 == 0:
        n = n//2
    else:
        n = 3*n+1
    
    result.append(n)

import matplotlib.pyplot as plt
plt.plot(result)
plt.xlabel('iteration',fontsize=15)
plt.ylabel('n',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
````
